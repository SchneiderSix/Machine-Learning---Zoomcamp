```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../datasets/jamb_exam_results.csv')
```

Prepare the data

```python
df.head()
```

```python
selected_columns = [x for x in df.columns if x != 'Student_ID']

df = df[selected_columns]

df.columns = df.columns.str.lower().str.replace(' ', '_')

df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.fillna(0, inplace=True)
df[
    df.parent_education_level.isnull()
].head()
```

```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)

df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train.jamb_score
y_val = df_val.jamb_score
y_test = df_test.jamb_score

del df_train['jamb_score']
del df_val['jamb_score']
del df_test['jamb_score']
```

```python
from sklearn.feature_extraction import DictVectorizer

train_dicts = df_train.to_dict(orient='records')

dv = DictVectorizer(sparse=True)

X_train = dv.fit_transform(train_dicts)
```

Question 1

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(
    max_depth=1
)

dt.fit(X_train, y_train)
```

```python
from sklearn.tree import export_text

print(export_text(dt, feature_names=dv.get_feature_names_out()))
```

study_hours_per_week is used for splitting the data

Question 2

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error

val_dicts = df_val.to_dict(orient='records')

dv = DictVectorizer(sparse=True)

X_val = dv.fit_transform(val_dicts)

rf = RandomForestRegressor(
    n_estimators=10,
    random_state=1,
    n_jobs=-1
)

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(
    X_val
)

rms = root_mean_squared_error(y_val, y_pred)

print(rms)
```

Question 3

```python
scores={}
for i in [10, 25, 80, 200]:
    rf = RandomForestRegressor(
        n_estimators=i,
        random_state=1,
        n_jobs=-1
    )

    rf.fit(X_train, y_train)

    y_pred = rf.predict(
        X_val
    )

    rms = root_mean_squared_error(y_val, y_pred)
    scores[i]=rms
print(scores)
```

at 80 n_estimators get the best rms

```python
import xgboost as xgb

features = list(dv.get_feature_names_out())
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
```

```python
def parse_xgb_output(output):
    results = []

    for line in output.stdout.strip().split('\n'):
        it_line, train_line, val_line = line.split('\t')

        it = int(it_line.strip('[]'))
        train = float(train_line.split(':')[1])
        val = float(val_line.split(':')[1])

        results.append((it, train, val))
    
    columns = ['num_iter', 'train_score', 'val_score']
    df_results = pd.DataFrame(results, columns=columns)
    return df_results
```
